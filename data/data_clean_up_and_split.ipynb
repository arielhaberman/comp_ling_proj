```python
#basically you need to change all the file paths because I was lazy but this should remove the nWord column from all the tsv files

def remove_nwords_column(input_folder, output_folder):
    # Create output directory if it does not exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Iterate over each file in the input folder
    for filename in os.listdir(input_folder):
        # Process only .tsv files
        if filename.endswith('.tsv'):
            file_path = os.path.join(input_folder, filename)
            # Load the TSV file into a DataFrame
            try:
                df = pd.read_csv(file_path, sep='\t')
                
                # Check if 'nWords' column exists and drop it if it does
                if 'nWords' in df.columns:
                    df = df.drop(columns=['nWords'])
                
                # Save the modified DataFrame to the output folder with the same filename
                output_path = os.path.join(output_folder, filename)
                df.to_csv(output_path, sep='\t', index=False)
                
                print(f"Processed {filename} - 'nWords' column removed.")
            except Exception as e:
                print(f"Error processing {filename}: {e}")

# Specify input and output folder paths
input_folder = 'C:/Users/ariel/Documents/second_year/comp_ling/ai-teacher-test/data/0_datasets/tscc_old/'
output_folder = 'C:/Users/ariel/Documents/second_year/comp_ling/ai-teacher-test/data/0_datasets/tscc/'

# Run the function
remove_nwords_column(input_folder, output_folder)
```

```python
#once again please change file paths but this is the train, test, split so dont change the random seed

import os
import shutil
import random

def split_dataset(input_folder, output_folder, train_ratio=0.75, dev_ratio=0.1, test_ratio=0.15):
    # Ensure the ratios add up to 1.0
    assert train_ratio + dev_ratio + test_ratio == 1.0, "Ratios must sum to 1.0"

    # Create subdirectories for train, dev, and test sets
    os.makedirs(os.path.join(output_folder, 'train'), exist_ok=True)
    os.makedirs(os.path.join(output_folder, 'dev'), exist_ok=True)
    os.makedirs(os.path.join(output_folder, 'test'), exist_ok=True)

    # List all TSV files in the input folder
    tsv_files = [f for f in os.listdir(input_folder) if f.endswith('.tsv')]
    random.shuffle(tsv_files)  # Shuffle the files for randomness

    # Calculate split sizes
    total_files = len(tsv_files)
    train_size = int(total_files * train_ratio)
    dev_size = int(total_files * dev_ratio)

    # Assign files to each set
    train_files = tsv_files[:train_size]
    dev_files = tsv_files[train_size:train_size + dev_size]
    test_files = tsv_files[train_size + dev_size:]

    # Move files to their respective directories
    for f in train_files:
        shutil.copy(os.path.join(input_folder, f), os.path.join(output_folder, 'train', f))
    for f in dev_files:
        shutil.copy(os.path.join(input_folder, f), os.path.join(output_folder, 'dev', f))
    for f in test_files:
        shutil.copy(os.path.join(input_folder, f), os.path.join(output_folder, 'test', f))

    print(f"Split complete: {len(train_files)} train, {len(dev_files)} dev, {len(test_files)} test files.")

# Usage
input_folder = 'C:/Users/ariel/Documents/second_year/comp_ling/ai-teacher-test/data/0_datasets/tscc_all/'
output_folder = 'C:/Users/ariel/Documents/second_year/comp_ling/ai-teacher-test/data/0_datasets/tscc_split/'
split_dataset(input_folder, output_folder, train_ratio=0.75, dev_ratio=0.1, test_ratio=0.15)
```
